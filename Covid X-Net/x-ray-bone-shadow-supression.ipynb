```python
def load_data(no_of_images):
    img_size = (224, 224)
    imgs_source = []
    imgs_target = []
    
    source_dir = "../input/xray-bone-shadow-supression/augmented/augmented/source"
    target_dir = "../input/xray-bone-shadow-supression/augmented/augmented/target"
    
    filenames = os.listdir(source_dir)  # List files directly instead of walking
    for i, filename in enumerate(filenames):
        if i >= no_of_images:  # Limit number of images
            break
        
        # Read and process source image
        img_source = cv2.imread(os.path.join(source_dir, filename), cv2.IMREAD_GRAYSCALE)
        img_source = cv2.resize(img_source, img_size)
        img_source = img_source / 255.0  # Normalize
        
        # Read and process target image
        img_target = cv2.imread(os.path.join(target_dir, filename), cv2.IMREAD_GRAYSCALE)
        img_target = cv2.resize(img_target, img_size)
        img_target = img_target / 255.0  # Normalize
        
        # Append to lists
        imgs_source.append(img_source)
        imgs_target.append(img_target)
    
    return np.array(imgs_source), np.array(imgs_target)
```

```python
source, target = load_data(1000)

print(f"Source images shape: {source.shape}")
print(f"Target images shape: {target.shape}")
```

### Reshaping and Spliting the Data

```python
IMG_HEIGHT = 224
IMG_WIDTH = 224
img_channels = 1  # Grayscale images
img_shape = (IMG_HEIGHT, IMG_WIDTH, img_channels)

# Reshaping source and target image.
source = np.array(source).reshape(-1, IMG_HEIGHT, IMG_WIDTH, img_channels)
target = np.array(target).reshape(-1, IMG_HEIGHT, IMG_WIDTH, img_channels)

# Splitting into train and test sets
from sklearn.model_selection import train_test_split
source_train, source_test, target_train, target_test = train_test_split(
    source, target, test_size=0.2, random_state=1
)


print(f"Train source shape: {source_train.shape}, Train target shape: {target_train.shape}")
print(f"Validation source shape: {source_test.shape}, Validation target shape: {target_test.shape}")
```

### Defining the Autoencoder Mode

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import RMSprop

def autoencoder(input_img):
    # encoder
    conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img) 
    conv1 = BatchNormalization()(conv1)
    conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(conv1)
    conv1 = BatchNormalization()(conv1)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1) 
    conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(pool1) 
    conv2 = BatchNormalization()(conv2)
    conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(conv2)
    conv2 = BatchNormalization()(conv2)
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2) 
    conv3 = Conv2D(128, (3, 3), activation='relu', padding='same')(pool2) 
    conv3 = BatchNormalization()(conv3)
    conv3 = Conv2D(128, (3, 3), activation='relu', padding='same')(conv3)
    conv3 = BatchNormalization()(conv3)

    # decoder
    conv4 = Conv2D(64, (3, 3), activation='relu', padding='same')(conv3) 
    conv4 = BatchNormalization()(conv4)
    conv4 = Conv2D(64, (3, 3), activation='relu', padding='same')(conv4)
    conv4 = BatchNormalization()(conv4)
    up1 = UpSampling2D((2, 2))(conv4) 
    conv5 = Conv2D(32, (3, 3), activation='relu', padding='same')(up1) 
    conv5 = BatchNormalization()(conv5)
    conv5 = Conv2D(32, (3, 3), activation='relu', padding='same')(conv5)
    conv5 = BatchNormalization()(conv5)
    up2 = UpSampling2D((2, 2))(conv5) 
    decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(up2) 
    return decoded

input_img = Input(shape=(224, 224, 1))  # Adjust shape as per your data
autoencoder_model = Model(input_img, autoencoder(input_img))
autoencoder_model.compile(loss='mean_squared_error', optimizer=RMSprop())

autoencoder_model.summary()
```

```python
n_epoch = 200
n_batch = 128

from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt

# Early stopping to avoid overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Learning rate scheduler to reduce learning rate on plateau
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)

# Model training with callbacks
autoencoder_train = autoencoder_model.fit(
    source_train, target_train,
    epochs=n_epoch,
    batch_size=n_batch,
    verbose=1,  # Set verbose to 1 for better monitoring
    validation_data=(source_test, target_test),
    callbacks=[early_stop, lr_scheduler]
)
```

```python
# Ensure interval_epochs are within the range of completed epochs
actual_epochs = len(autoencoder_train.history['loss'])  # Get the actual number of completed epochs
interval_epochs = [e for e in [0, 49, 99, 149, 199] if e < actual_epochs]  # Filter out invalid indices

# Print loss and validation loss for valid epochs
for e in interval_epochs:
    print("epoch = {}\tLoss = {:.5f}\tValidation_Loss = {:.5f}".format(
        e + 1,
        autoencoder_train.history['loss'][e],
        autoencoder_train.history['val_loss'][e]
    ))
```

```python
# Plotting training and validation loss
plt.plot(autoencoder_train.history['loss'], label='Training Loss')
plt.plot(autoencoder_train.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Function to predict and visualize an image with confidence and class
def predict_image(model, img_path, target_size):
    """
    Predict the reconstructed output for a given image using a trained autoencoder.
    """
    # Load and preprocess the image
    img = load_img(img_path, color_mode="grayscale", target_size=target_size)
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array = img_array / 255.0  # Normalize pixel values

    # Perform prediction
    reconstructed_img = model.predict(img_array)
    difference = np.abs(img_array - reconstructed_img)  # Calculate difference as proxy for confidence
    confidence = 1 - np.mean(difference)

    # Display the original and reconstructed image
    plt.figure(figsize=(8, 4))

    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(img_array.squeeze(), cmap='gray')  # Use img_array here
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title(f"Predicted Image\nConfidence: {confidence:.2f}")
    plt.imshow(reconstructed_img.squeeze(), cmap='gray')
    plt.axis("off")

    plt.tight_layout()
    plt.show()

    return confidence
```
